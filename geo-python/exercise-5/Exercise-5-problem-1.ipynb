```python
import pathlib 
NOTEBOOK_PATH = pathlib.Path().resolve()
DATA_DIRECTORY = NOTEBOOK_PATH / "data"
OUTPUT_DIRECTORY = NOTEBOOK_PATH / "docs"
```

```python
import geopandas 
import matplotlib.pyplot as plt

map_data_path = '../exercise-4/data/grid.gpkg'
data = geopandas.read_file(map_data_path)
road_data = geopandas.read_file('./data/Tieosoiteverkko-2025-01-01.zip')

ax = data.plot(
    figsize=(12, 8),
    column="closest_time",
    scheme="quantiles",
    cmap="Spectral",
    linewidth=0,
    alpha=0.8
)

road_data[road_data.AJORATA>0].plot(column='AJORATA', ax=ax)
bounds = data.total_bounds
plt.xlim((bounds[0], bounds[2]))
plt.ylim((bounds[1], bounds[3]))

# Map overlaying travel time to supermarkets by public transport with motorways
plt.savefig('./data/static-map.png')
```
